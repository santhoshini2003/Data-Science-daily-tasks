```python
import numpy as np
import pandas as pd
data=pd.read_csv("House Price India.csv")
data
```

```python
skew_value = data["Price"].skew()
print("\nSkewness Value :", round(skew_value, 4))
```

```python
if skew_value > 0:
    print("positive skew")
elif skew_value < 0:
    print("negative skew")
else:
    print("normal skew")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(data["Price"], bins=6)
plt.title("House Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
