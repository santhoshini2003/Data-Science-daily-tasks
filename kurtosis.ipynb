```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")
df
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
price=df['PRICEEACH']
```

```python
kurtosis_value=price.kurt()
```

```python
print("kurtosis_value:",kurtosis_value)

#validation
if kurtosis_value > 0:
    print("Leptokurtic Distribution")
elif kurtosis_value < 0:
    print("Platykurtic Distribution")
else:
    print("Mesokurtic Distribution")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(price, bins=10)
plt.title("sales data PRICEEACH")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sales=df['SALES']
```

```python
kurtosis_value=sales.kurt()
```

```python
print("kurtosis_value:",kurtosis_value)

#validation
if kurtosis_value > 0:
    print("Leptokurtic Distribution")
elif kurtosis_value < 0:
    print("Platykurtic Distribution")
else:
    print("Mesokurtic Distribution")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(sales, bins=10, color="red", edgecolor="black")
plt.title("sales data SALES")
plt.xlabel("sales")
plt.ylabel("Frequency")
plt.show()
```
