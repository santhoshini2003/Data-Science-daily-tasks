```python
from scipy.stats import norm
#given values
mu=100
sigma=15
x=90-110
prob_greater=1-norm.cdf(x,mu,sigma)
print("probability of values:",round(prob_greater * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#normal distribution parameters
mu=100
sigma=15

#create random sample dataset
x=np.random.normal(mu,sigma, 10000)

#histogram (bell shape)
plt.hist(x,bins=30,density=True,alpha=0.6)

#plot normal distribution curve
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax, 1000)
y=norm.pdf(x,mu,sigma)
plt.plot(x,y,linewidth=2)

#label and title
plt.title("Normal Distribution Histogram\nBell Shaped Curve")
plt.xlabel("height(cm)")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

```python
#Exponential distribution
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

#given values
mean_time=10
t=3

lambda_value=1/mean_time
prob_within=expon.cdf(t,scale=1/lambda_value)

print("Probability customers arrives within 3 minutes:",round(prob_within*100,2),"%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
mean_time=10 #average customers arrival time

#generate random exponentail data
data = np.random.exponential(mean_time, 10000)

#historgram
plt.hist(data,bins=30,density=True)
plt.title("Exponential Distribution - Histogram")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
