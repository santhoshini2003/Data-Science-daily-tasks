```python
df=df.fillna("Unknown")
print(df)
```

```python
import matplotlib.pyplot as plt
plt.bar(
    df["Product_Category"],
    df["Sales_Amount"],
)
plt.title("Sales_Amount and Product_Category")
plt.xlabel("Sales_Amount")
plt.ylabel("Product_Category")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.pie(
    df["Sales_Amount"],
    labels=df["Product_Category"],
    autopct="%10.10f%%"
)
plt.title("Sales Amount by Product Category")
plt.show()
```
